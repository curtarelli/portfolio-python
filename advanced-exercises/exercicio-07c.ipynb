# Exercício 1

```python
# Gráfico de colunas para os dados da população feminina
x_pos = [x for x in range(len(faixa_etaria))]

plt.figure(figsize=(8, 12))

plt.bar(x_pos, feminino, align='center',
       color='crimson', linewidth=2, edgecolor='black')

plt.yticks([0, 100000, 500000, 1000000, 2000000, 3000000, 4000000, 5000000, 6000000, 7000000, 8000000, 9000000],
          ['0', '0,1 mi', '0,5 mi', '1 mi', '2 mi', '3 mi', '4 mi', '5 mi', '6 mi', '7 mi', '8 mi', '9 mi'])

plt.xticks(x_pos, faixa_etaria, rotation=45)

plt.title('População Feminina por Faixa Etária')

plt.xlabel('Faixa etária')
plt.ylabel('População')

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5);
```

```python
# Gráfico de colunas para os dados da população feminina
y_pos = [x for x in range(len(faixa_etaria))]

plt.figure(figsize=(18, 10))

plt.barh(y_pos, masculino, align='center',
       color='blue', linewidth=2, edgecolor='black')

plt.xticks([0, 500000, 1000000, 2000000, 3000000, 4000000, 5000000, 6000000, 7000000, 8000000, 9000000],
          ['0', '0,5 mi', '1 mi', '2 mi', '3 mi', '4 mi', '5 mi', '6 mi', '7 mi', '8 mi', '9 mi'])

plt.yticks(y_pos, faixa_etaria, rotation=45)

plt.title('População Masculina por Faixa Etária')

plt.xlabel('População')
plt.ylabel('Faixa etária')

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5);
```
