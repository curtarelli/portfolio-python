```python
# Usando o matplotlib e numpy
t = np.linspace(0.0, 2.0, 100)

f = np.sin(np.pi * t)

plt.figure(figsize=(10, 8))

plt.plot(t,f, 'orange', linewidth=2.0)

plt.xlim([0.0, 2.0])
plt.ylim([-1.0, 1.0])

plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$f(t)$', fontsize=16)

plt.title(r'$f(t)=sin(\pi \times t)$', fontsize=24)

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5);
```

```python
# Vários gráficos em uma mesma área 1
t = np.linspace(0.0, 2.0, 100)

f1 = np.sin(np.pi * t)
f2 = np.sin(np.pi * 2 * t)

plt.figure(figsize=(10, 8))

plt.plot(t, f1, 'orange', linewidth=2.0,
        label=r'$f_1(t)=sin(\pi \times t)$')

plt.plot(t, f2, 'purple', linewidth=2.0,
        label=r'$f_2(t)=sin(2 \times \pi \times t)$')

plt.xlim([0.0, 2.0])
plt.ylim([-1.0, 1.0])

plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$f(t)$', fontsize=16)

plt.title('Senóides', fontsize=24)

plt.legend()

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5);
```

```python
# Vários gráficos em uma mesma área 2
t = np.linspace(0.0, 2.0, 100)

f1 = np.sin(np.pi * t)
f2 = np.sin(np.pi * 2 * t)

plt.figure(figsize=(10, 8))

plt.plot(t, f1, 'orange',
         t, f2, 'purple',
         linewidth=2.0)

plt.xlim([0.0, 2.0])
plt.ylim([-1.0, 1.0])

plt.xlabel(r'$t$', fontsize=16)
plt.ylabel(r'$f(t)$', fontsize=16)

plt.title('Senóides', fontsize=24)

plt.legend([r'$f_1(t)=sin(\pi \times t)$',
           r'$f_2(t)=sin(2 \times \pi \times t)$'])

plt.grid(b=True, color='gray', linestyle='--', linewidth=0.5);
```

```python
# Vários gráficos em uma mesma figura, multiplas áreas de desenho
t = np.linspace(0.0, 2.0, 100)

f1 = np.sin(np.pi * t)
f2 = np.sin(np.pi * 2 * t)
f3 = np.sin(np.pi * 4 * t)
f4 = np.sin(np.pi * 8 * t)

plt.figure(figsize=(10, 8))

plt.subplot(221, facecolor='#FFFFEA')
plt.plot(t, f1, 'orange')
plt.title(r'$f_1(t)=sin(\pi \times t)$')

plt.subplot(222)
plt.plot(t, f2, 'purple')
plt.title(r'$f_2(t)=sin(2 \times \pi \times t)$')

plt.subplot(223)
plt.plot(t, f3, 'salmon')
plt.title(r'$f_2(t)=sin(4 \times \pi \times t)$')

plt.subplot(224, facecolor='#FEEEEE')
plt.plot(t, f4, 'crimson')
plt.title(r'$f_2(t)=sin(8 \times \pi \times t)$')

plt.suptitle('Senóides', fontsize=24);
```

```python
# Scatter plot para N valores com coordenadas x e y e peso (áreas), onde o peso define o tamanho da figura no gráfico 
N = 80

# vamos sortear N valores no intervalo [0, 1) para cada eixo.
x = np.random.rand(N)
y = np.random.rand(N)

# vamos sortear N valores no intervalo [0, 1) para associar a cores.
colors = np.random.rand(N)

# cada elemento será um círculo de raio entre 0 e 20 pontos.
area = np.pi * (20 * np.random.rand(N))**2

plt.figure( figsize=(10, 8) )

plt.scatter(x, y, s=area, c=colors, alpha=0.5);
```

```python
# Histograma simples
μ = 200
σ = 25

values = np.random.normal(μ, σ, size=100)

plt.figure( figsize=(8, 8) )

plt.hist(values, bins=10, density=False, align='left',
         facecolor='green', edgecolor='black', linewidth=1.2, alpha=0.65);

plt.title(r"Distribuição Normal: $\mu=200$ e $\sigma=25$")

plt.xlabel("Valor")
plt.ylabel("Frequência");
```

```python
# Histograma com curva de distribuição normal
μ = 200
σ = 25

values = np.random.normal(μ, σ, size = 100)

plt.figure( figsize=(15, 5) )

count, bins, ignored = plt.hist(values, bins=30, density=True,
                                align='left', facecolor='lightblue',
                                edgecolor='black', linewidth=1.2,
                                alpha=0.65);

plt.title(r"Distribuição Normal: $\mu=200$ e $\sigma=25$")

plt.xlabel("Valor")
plt.ylabel("Frequência")

plt.plot( bins, 1 / ( σ * np.sqrt(2 * np.pi) ) *
                np.exp( - (bins - μ)**2 / (2 * σ**2) ),
          linewidth=1.5, color='r' );
```
